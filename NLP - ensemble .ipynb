```python
import pandas as pd
data = pd.read_csv('./clean_data_am_nam_file/word2vec_features_with_labels_after_removing_noisy_columns.csv')
```

```python
data = data.transpose()
data
```

```python
# Find the top 30 words said by each piece of news
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
top_dict = {}
for c in data.columns:
    data[c] = le.fit_transform(data[c].astype(str))
    top = data[c].sort_values(ascending=False).head(30)
    top_dict[c]= list(zip(top.index, top.values))

top_dict
```

```python
# Print the top 15 words said by each piece of news
for news, top_words in top_dict.items():
    print(news)
    print(', '.join([word for word, count in top_words[0:14]]))
    print('---')
```

```python
# Look at the most common top words --> add them to the stop word list
from collections import Counter

# Let's first pull out the top 30 words for each piece of news
words = []
for news in data.columns:
    top = [word for (word, count) in top_dict[news]]
    for t in top:
        words.append(t)
        
words
```

```python
# Let's aggregate this list and identify the most common words along with how many routines they occur in
Counter(words).most_common()
```

DT,KNN,RF,XGB models into emsemble

```python
data_training = pd.read_csv('./clean_data_am_nam_file/word2vec_features_with_labels_after_removing_noisy_columns.csv')
```

```python
import numpy as np
np.random.seed(1110)
data_withoutlabel = data_training.drop(data_training.iloc[:,-2:], axis=1)
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val = train_test_split(data_withoutlabel,data_training.iloc[:,-2],test_size=0.2)
```

```python
# If  using xgb then need to transform labels into binary variables,but using XGBclassifier we don't have to
#from sklearn import preprocessing
#le = preprocessing.LabelEncoder()
#y_train1 = le.fit_transform(y_train)
#y_val1 = le.fit_transform(y_val)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
DT = DecisionTreeClassifier()
RF = RandomForestClassifier()
KNN = KNeighborsClassifier()
y_pred_dt = DT.fit(X_train, y_train).predict(X_val)
y_pred_rf = RF.fit(X_train, y_train).predict(X_val)
y_pred_knn = KNN.fit(X_train, y_train).predict(X_val)
print("Number of mislabeled points out of a total %d points : %d"
      % (X_val.shape[0], (y_val != y_pred_dt).sum()))
print("Number of mislabeled points out of a total %d points : %d"
      % (X_val.shape[0], (y_val != y_pred_rf).sum()))
print("Number of mislabeled points out of a total %d points : %d"
      % (X_val.shape[0], (y_val != y_pred_knn).sum()))
```

```python
from xgboost import XGBClassifier
XGB = XGBClassifier()
y_pred_xgb = XGB.fit(X_train, y_train).predict(X_val)
print("Number of mislabeled points out of a total %d points : %d"
      % (X_val.shape[0], (y_val != y_pred_xgb).sum()))
```

```python
print(f"Accuracy of DT is {DT.score(X_val, y_val)*100}%")
print(f"Accuracy of KNN is {RF.score(X_val, y_val)*100}%")
print(f"Accuracy of RF is {KNN.score(X_val, y_val)*100}%")
print(f"Accuracy of XGB is {XGB.score(X_val, y_val)*100}%")
```

```python
from sklearn.ensemble import VotingClassifier
# set equal weights for each of the classifiers to reproduce the basic majority vote ensemble:
model = VotingClassifier(estimators=[('dt', DT), ('knn', KNN), ('rf', RF),('xgb',XGB)], voting='hard')


y_pred_ensemble = model.fit(X_train, y_train).predict(X_val)

print(f"Accuracy of ensemble {model.score(X_val, y_val)*100}%")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val,y_pred_ensemble))
```
